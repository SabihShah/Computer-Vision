```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('image.jpg')
print(img.dtype)
height, width, _ = img.shape

cv2.imshow("Image", img)
cv2.waitKey()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# cv2.imshow("Gray", gray)
# cv2.waitKey()
```

```python
blur = cv2.GaussianBlur(gray, (15, 15), 1)
# cv2.imshow("Blur", blur)
# cv2.waitKey()
```

```python
edge = cv2.Canny(blur, 50, 150)
# cv2.imshow("Edge", edge)
# cv2.waitKey()
```

```python
lines = cv2.HoughLines(edge, 1, np.pi/180, 150)
print(lines.shape)
print(lines)
for line in lines:

    rho,theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a*rho
    y0 = b*rho
    x1 = int(x0 + 400*(-b))
    y1 = int(y0 + 400*(a))
    x2 = int(x0 - 400*(-b))
    y2 = int(y0 - 400*(a))

    cv2.line(img,(x1,y1),(x2,y2),(0,255,0),1)
```

```python
lines = cv2.HoughLinesP(edge, 1, np.pi/180, 30, maxLineGap=250)      # Return coordinates of the detected liens
print(lines.shape, lines)
for line in lines:
   x1, y1, x2, y2 = line[0]
   cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 1)
```
